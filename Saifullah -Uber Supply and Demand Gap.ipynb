```python
#importing relevant files
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
#reading the csv file into df
df = pd.read_csv(r"H:\UpGrad\Statistics & EDA\Ubber Supply-Demand Gap\Uber Request Data.csv")
#displaying the data frame
df
```

```python
#Unifying the date-time format
df['Request timestamp'] = pd.to_datetime(df['Request timestamp'])
df['Drop timestamp'] = pd.to_datetime(df['Drop timestamp'])
df
```

```python
#separating date and hour column
df['Request Hour'] = df['Request timestamp'].dt.hour
df['Drop hour'] = df['Drop timestamp'].dt.hour
df
```

```python
#Derive a day of the week column from Request timestamp column
df['Day'] = df['Request timestamp'].dt.day_name()
df
```

```python
#looking for the number of null values in each column
print(df.isnull().sum())
```

```python
# Divide the column of time into various period
df = df.assign(session=pd.cut(df['Request timestamp'].dt.hour,[0,4,7,10,15,18,20,24],labels=['LateNight','Early Morning','Morning','Afternoon','Evening','Late Evening','Night']))
df
```

```python
#creating a new data frame with status only
af = df[['Status']]
af
```

```python
#Calculating the frequency of the status
af['freq'] = af.groupby('Status')['Status'].transform('count')
af
```

```python
#Dropping the duplicate files
af.drop_duplicates(keep ='first',inplace =True)
af
```

```python
#PLotting the frequency of request that gets cancelled
plt.pie(af['freq'], labels=af['Status'],autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Frequency of each type of trip")
plt.show()
```

```python
#Creating a new data frame for the Pick up point
fd = df[['Status','Pickup point']]
fd
```

```python
# Remove the rows where status is trip completed
fd = fd.drop(fd[fd['Status'] == 'Trip Completed'].index)
fd
```

```python
#calculating the frequency of each pickup point and groupin by pickup point and status
fd = fd.groupby(['Pickup point','Status']).size().reset_index(name = 'Freq')
fd
```

```python
#Grouping by pivot table and plotting it
fd.pivot("Pickup point", "Status", "Freq",).plot.bar(stacked = True)
plt.title("Frequency of incomplete ride by location subdivided")
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
#Creating a new data frame for the Hour of the day point
hr = df[['Status','session']]
hr
```

```python
# Remove the rows where status is trip completed
hr = hr.drop(hr[hr['Status'] == 'Trip Completed'].index)
hr
```

```python
#calculating the frequency of each pickup point and groupin by pickup point and status
hr = hr.groupby(['session','Status']).size().reset_index(name = 'Freq')
hr
```

```python
#Grouping by pivot table and plotting it
hr.pivot("session", "Status", "Freq",).plot.bar(stacked = True)
plt.title("Frequency of incomplete ride by Hour of day")
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
#Creating a new data frame for the Hour of the day and pickup point 
pr = df[['Status','Pickup point','session']]
pr
```

```python
# Remove the rows where status is trip completed
pr = pr.drop(pr[pr['Status'] == 'Trip Completed'].index)
pr
```

```python
#calculating the frequency of each pickup point and groupin by pickup point and session
pr = pr.groupby(['Pickup point','session','Status']).size().reset_index(name = 'Freq')
pr
```

```python
# Selecting airport pickup point data separately
air = pr[pr['Pickup point']== 'Airport']
air
```

```python
#Dropping the Pickup point column
air.drop('Pickup point',axis =1,inplace =True)
air
```

```python
#Grouping by pivot table and plotting it
air.pivot("session","Status", "Freq").plot.bar(stacked = True)
plt.title("Frequency of incomplete ride from airport to city by Hour of the day")
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1))
plt.show()
```

```python
# Selecting city pickup point data separately
city = pr[pr['Pickup point']== 'City']
city
```

```python
#Dropping the Pickup point column
city.drop('Pickup point',axis =1,inplace =True)
city
```

```python
#Grouping by pivot table and plotting it
city.pivot("session","Status", "Freq").plot.bar(stacked = True)
plt.title("Frequency of incomplete ride from city to airport by Hour of the day")
plt.legend(loc='upper center', bbox_to_anchor=(1.2, 1))
plt.show()
```
